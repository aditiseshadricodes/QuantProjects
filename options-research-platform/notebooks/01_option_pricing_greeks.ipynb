Test whether the data loader can fetch available option expiries and a single option chain for one liquid ticker.

```python
from pathlib import Path
import sys
import pandas as pd
import os
import matplotlib.pyplot as plt
import json

from dotenv import load_dotenv

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.data_loader import get_option_expiries, fetch_option_chain, fetch_option_chains_for_universe
```

```python
ticker = 'AAPL'
expiries = get_option_expiries(ticker)
print(len(expiries))
expiry = expiries[0]
chain = fetch_option_chain(ticker, expiry)
print(chain.shape)
print(chain.value_counts())
print(chain.columns)
```
